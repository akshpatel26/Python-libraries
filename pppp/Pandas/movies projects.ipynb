```python
top_movie = data.loc[data['Rating'].sort_values(ascending=False)[:12].index]
top_movie
```

```python
sns.histplot(data=top_movie, x="Year",hue="Rating", multiple="stack")
plt.title('Distribution of Top 10 Movies', fontsize=16)
plt.xlabel('Year', fontsize=9)
plt.ylabel('Rating', fontsize=9)
plt.tight_layout()
plt.show()
```

```python
genre_counts = data['Genre'].value_counts().reset_index()
genre_counts.columns = ['Genre', 'Count']

# Select the top N genres (e.g., top 5)
top_n_genres = genre_counts.head(7)
top_n_genres
```

```python
plt.figure(figsize=(4, 4))
ex=[0.1,0,0,0,0,0,0]
plt.pie(top_n_genres['Count'], labels=top_n_genres['Genre'], autopct='%1.1f%%', startangle=140,explode=ex)
sns.color_palette(palette='Dark2')

plt.title('Movie Genre Distribution', fontsize=16)
plt.axis('equal')
plt.show()

```

```python
# Group the data by director and calculate the average rating
director_avg_rating = data.groupby('Director')['Rating'].mean().reset_index()

director_avg_rating = director_avg_rating.sort_values(by='Rating', ascending=False)

top_directors = director_avg_rating.head(7)
top_directors
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=top_directors, x='Rating', y='Director', palette='Dark2')

plt.title('Top Directors by Average Rating', fontsize=16)
plt.xlabel('Average Rating', fontsize=14)
plt.ylabel('Director', fontsize=14)

plt.show()
```

```python
actor_counts = data['Actor 1'].value_counts().reset_index()
actor_counts.columns = ['Actor', 'MovieCount']

top_n_actors = actor_counts.head()
top_n_actors
```

```python
plt.figure(figsize=(7, 4))
sns.barplot(data=top_n_actors, x='MovieCount', y='Actor', orient='h')

# Set plot labels and title
plt.title('Top Actors by Number of Movies', fontsize=16)
plt.xlabel('Number of Movies', fontsize=14)
plt.ylabel('Actor', fontsize=14)

# Show the plot
plt.show()
```

```python
year_movie_counts = data['Year'].value_counts().reset_index()
year_movie_counts.columns = ['Year', 'MovieCount']

year_movie_counts = year_movie_counts.sort_values(by='Year')
year_movie_counts
```

```python
plt.figure(figsize=(7, 4))
sns.plot(data=year_movie_counts, x='Year', y='MovieCount')

plt.title('Number of Movies Released Every Year', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Movies Released', fontsize=14)

plt.show()
```

```python
filtered_df = data[(data['Rating'] > 8) & (data['Votes'] >'70000')]
filtered_df.head(10)
```

```python

plt.figure(figsize=(10, 6))
ax=sns.barplot(data=filtered_df,x='Name',y='Votes',hue='Rating',dodge=False,width=0.8,palette='muted')

ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha='right')
ax.legend(loc='upper right')
ax.set_xlabel('Movie Name')
ax.set_ylabel('Votes')
ax.set_title('Movies with rating greater than 8 and votes greater than 10000')

plt.show()
     
```
